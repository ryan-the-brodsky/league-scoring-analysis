```python
def clean_position(position):
    position = str(position)
    last_two_letters = position[-2:len(position)].upper()
    if last_two_letters == "LB":
        return "LB"
    elif last_two_letters in ["SS", "CB", "FS", "DB"]:
        return "DB"
    elif last_two_letters in ["DE", "DT", "NT"]:
        return "DL"
    elif last_two_letters == "AN":
        return None
    else:
        return last_two_letters
```

```python
# At this point we have all_defensive_players and all_offensive_players comprising three seasons of stats
# map(scoring_system.keys(), lambda scoring_key : player[scoring_key] * scoring_system[scoring_key]
def tally_points_for_player(player):
    total = 0
    for scoring_key in scoring_system.keys():
        if scoring_key in player:
            total += scoring_system[scoring_key] * player[scoring_key]
    return total

all_defensive_players['fantasy_pts'] = all_defensive_players.apply(tally_points_for_player, axis=1)
all_offensive_players['fantasy_pts'] = all_offensive_players.apply(tally_points_for_player, axis=1)
tes = all_offensive_players[all_offensive_players['position'] == "TE"].sort_values(by="fantasy_pts", ascending=False)
tes[['Player', 'Season', 'fantasy_pts']].head(108)
```

```python
# Convention Day Helper functions
def display_descriptive_statistics_for_top_x_players(x):
    ultimate_data = []
    index = 0
    for position in positions:
        nums = position.sort_values(by="fantasy_pts", ascending=False).head(x)["fantasy_pts"].describe().to_dict()
        nums['position'] = position_names[index]
        index += 1
        ultimate_data.append(nums)
    ultimate_df = pd.DataFrame(ultimate_data)
    ultimate_df.set_index("position", inplace=True)
    ultimate_df['First2Worst'] = ultimate_df['max'] - ultimate_df['min']
    ultimate_df['First2Avg'] = ultimate_df['max'] - ultimate_df['mean']
    ultimate_df['First2Worst_perG'] = ultimate_df['First2Worst']/16.5
    ultimate_df['First2Avg_perG'] = ultimate_df['First2Avg']/16.5
    return ultimate_df
    
def display_top_x_players_at_position(x, pos):
    players = positions[position_names.index(pos)]
    results = players.sort_values(by="fantasy_pts", ascending=False)[['Player', "Season", "fantasy_pts"]].head(x).reset_index(drop=True)
    results.index += 1
    return results
    
def display_top_x_players_at_y_position_in_z_year(x, y, z):
    players = positions[position_names.index(y)]
    results = players[players['Season'] == z].sort_values(by="fantasy_pts", ascending=False)[['Player', "Season", "fantasy_pts"]].head(x).reset_index(drop=True)
    results.index += 1
    return results

# for clarity for you Ryan, reset index goes to 0-based & no direct options to start at 1
# so I'm saving the results to a variable than incrementing all indices before returning
```
